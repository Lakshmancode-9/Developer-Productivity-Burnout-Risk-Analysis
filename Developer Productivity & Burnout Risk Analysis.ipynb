```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("default")
```

```python
df = pd.read_csv("Technology_and_AI_Companies_Market_Dataset.csv")
df.head()
```

```python
print("Rows, Columns:", df.shape)
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df = df.dropna() #Drops if any missing values exist
```

```python
#EXPLORATORY DATA ANALYSIS (EDA)
```

```python
#Market Capitalization by Sector
sector_cap = df.groupby("Sector")["Market_Cap_Trillion_USD"].sum().sort_values(ascending=False)

plt.figure(figsize=(10,5))
sector_cap.plot(kind="bar")
plt.title("Market Capitalization by Technology Sector")
plt.xlabel("Sector")
plt.ylabel("Market Cap (Trillion USD)")
plt.show()
```

```python
#Top 10 Companies by Revenue
top_revenue = df.sort_values("Revenue_2024_Billion_USD", ascending=False).head(10)
plt.figure(figsize=(10,5))
sns.barplot(data=top_revenue, x="Company", y="Revenue_2024_Billion_USD")
plt.xticks(rotation=45)
plt.title("Top 10 AI Companies by Revenue (2024)")
plt.show()
```

```python
#AI Revenue Dependency by AI Focus Area
plt.figure(figsize=(10,5))
sns.lineplot(
    data=df,
    x="AI_Focus_Area",
    y="AI_Revenue_Percent",
    marker="o"
)
plt.xticks(rotation=45)
plt.title("AI Revenue Dependency by AI Focus Area")
plt.show()
```

```python
#R&D Spending vs AI Patents ==> Innovation Efficiency
plt.figure(figsize=(8,6))
sns.scatterplot(
    data=df,
    x="R&D_Spending_Billion_USD",
    y="AI_Patents_Count",
    size="Market_Cap_Trillion_USD",
    hue="Sector",
    alpha=0.7
)
plt.title("Innovation Efficiency: R&D vs AI Patents")
plt.show()
```

```python
#Growth Rate vs Stock Price
plt.figure(figsize=(8,6))
sns.scatterplot(
    data=df,
    x="Growth_Rate_2024_Percent",
    y="Stock_Price_USD",
    size="Revenue_2024_Billion_USD",
    alpha=0.7
)
plt.title("Growth Rate vs Stock Price")
plt.show()
```

```python
#ESG Score vs AI & Cloud Revenue
esg_group = df.groupby("ESG_Score")[["AI_Revenue_Percent", "Cloud_Revenue_Billion_USD"]].sum()
esg_group.plot(kind="bar", figsize=(8,5))
plt.title("AI & Cloud Revenue by ESG Score")
plt.show()
```

```python
#Distribution of AI Revenue Dependency ==> Histogram
plt.figure(figsize=(8,5))
plt.hist(df["AI_Revenue_Percent"], bins=10)
plt.title("Distribution of AI Revenue Percentage")
plt.xlabel("AI Revenue %")
plt.ylabel("Company Count")
plt.show()
```

```python
#AI Focus Area vs Growth Rate
focus_growth = df.groupby("AI_Focus_Area")["Growth_Rate_2024_Percent"].mean().sort_values()
plt.figure(figsize=(10,5))
focus_growth.plot(kind="barh")
plt.title("Average Growth Rate by AI Focus Area")
plt.xlabel("Growth Rate (%)")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(
    df.select_dtypes(include=np.number).corr(),
    cmap="coolwarm",
    annot=False
)
plt.title("Correlation Heatmap")
plt.show()
```

```python
print("Key Insights:")
print("- AI market value is concentrated in Semiconductor & Software sectors.")
print("- Innovation efficiency varies significantly across companies.")
print("- ESG performance aligns with sustainable AI revenue growth.")
print("- Growth rate alone does not determine stock valuation.")
```
